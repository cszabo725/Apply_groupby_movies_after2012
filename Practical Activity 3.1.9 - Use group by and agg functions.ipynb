```python
# Import the required packages
import pandas as pd

# Create the DataFrames
movies = pd.read_excel('movies_merge.xlsx')
ott = pd.read_csv('ott_merge.csv')

#View the DataFrame
movies.head()
```

```python
# Merge the DataFrames with a left join 
Movies_L = movies.merge(ott, how='left', on='ID') 

# View the Movies_L DataFrame
Movies_L
```

```python
# Merge the DataFrames with a righ join 
Movies_R = movies.merge(ott, how='right', on='ID') 

# View the Movies_L DataFrame
Movies_R
```

```python
# Number of movies on Netflix by year, 2012 and after
mo_gpby=Movies_L.groupby(['Year'])['Netflix'].agg('sum').reset_index(0)
mo_gpby[mo_gpby['Year'] >= 2012]
```

```python
# Movie runtime by year, 2012 and after 
mo_gpby1=Movies_L.groupby(['Year'])['Runtime'].agg('sum').reset_index()
mo_gpby1[mo_gpby1['Year'] >= 2012]
```

```python
# Determing worst review score via Rotten Tomatoes, 2012 and after
mo_gpby2 = Movies_L.groupby(['Year'])['Rotten Tomatoes'].agg('min').reset_index()
mo_gpby2[mo_gpby2['Year'] >= 2012]
```

```python
# Determing best review score via Rotten Tomatoes, 2012 and after
mo_gpby2 = Movies_L.groupby(['Year'])['Rotten Tomatoes'].agg('max').reset_index()
mo_gpby2[mo_gpby2['Year'] >= 2012]
```
